### PANDAS 

To get all the data in Pythhon from the SQL database FOR loops will be used to create a dataframe for each table. This will make it easier to analyse the useful data and maniplutae it. 

```python
# get the names of the tables and store it in an array 
data_names = []
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
for line in cur.fetchall():
    data_names.append(line[0])

# make an array of the names the data frames should be    
data_frame_names = [] 

for j in data_names : 
    data_frame_names.append(j+'_df')
data_frame_names


# get data from the data base 
for x,n in enumerate(data_names) : 

    format_str1 = "SELECT * FROM {name};"
    sql_command1 = format_str1.format(name = n)
    
    cur.execute(sql_command1)
    exec(data_frame_names[x] + '= pd.DataFrame(cur.fetchall())')

# get column name values
col = list(range(len(data_names)))

for x,n in enumerate(data_names) : 
    
    col_names = []
    format_str2 = "PRAGMA table_info({name});"
    sql_command2 = format_str2.format(name = n)
    cur.execute(sql_command2)
    for line in cur.fetchall():
        col_names.append(line[1])
    col[x] = list(col_names)

# assign column names to data frames 

for x,i in enumerate(col) :
    # check to make sure the data_frames are not empty to avoid errors 
    if (eval(data_frame_names[x])).columns.size == 0:
        pass
    else : 
        (eval(data_frame_names[x])).columns = i
```

### Heat Map 

```python
# Create a new df called "dual_types" by getting rid of the records for which we do not have Type 2

dual_types = pokedex[pokedex['type2'] != 'NoneType']

# create a grid from the dual_types dataframe 

grid = dual_types.groupby(['type1','type2']).size()
grid = grid.unstack()
grid
```

```python
# display the grid in a heat map 

sb.set(rc={'figure.figsize':(14,10)})

sb.heatmap(grid,cmap="RdPu",annot=True,linewidths=.75)
```

It can be seen that the the most common type of dual pokemon is type1 normal and type 2 flyging. 

There are also alot of type 1 grass and type 2 poison and type 1 bug and type 2 flying. 

It can also be seen that the variety of dual pokemon is quite diverse and the majority of unique combinations of types 1 and 2  pokemon have a count of 1 to 4. 

To get a more accurate representation a different metric will be used. The six statistic columns wil be standardized  independently by converting each value into a z-score so when we do take the sum, we account for the variation in the each statistic using its mean and standard deviation across all Pokemon species. 



```python
# create names for z-stats table 
z_names = ['id_z','hp_z','attack_z','defense_z','sp_atk_z','sp_def_z','speed_z','total_z']

# get the z score for every column 

for x,i in enumerate(poke_statistics):
    z = (poke_statistics[i] - np.mean(poke_statistics[i]))/(np.std(poke_statistics[i]))
    exec(z_names[x] + " =z")
    
# create data frame 

z_stats = pd.DataFrame({'hp_z' : hp_z,
                        'attack_z' : attack_z,
                        'defense_z' : defense_z, 
                        'sp_atk_z': sp_atk_z,
                        'sp_def_z' : sp_def_z,
                        'speed_z' : speed_z,
})

z_stats.head()
```

```python
# define a new colum strength to sum the overall z scores to find the strongest pokemon 
pokedex['strength'] = z_stats.sum(axis=1)
df.head()
```

```python


# Repeat the heatmap you created but now for the new strength metric you computed
dual_types_z = pokedex[pokedex['type2'] != 'nothing']

# create a grid using dual types df 
grid_z = dual_types_z.groupby(['type1','type2']).mean()['strength']
grid_z = grid_z.unstack()

sb.set(rc={'figure.figsize':(14,10)})

sb.heatmap(grid_z,cmap="RdPu",annot=True,linewidths=.75)
```
